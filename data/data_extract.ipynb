```python
import pandas as pd
import numpy as np

# put original data
climate_df = pd.read_csv('./data/climate_all.csv') 


if climate_df is not None:
    # --- step 1: define constants ---
    # random seed for reproducibility
    RANDOM_SEED = 42
    SAMPLE_SIZE = 10

    # --- step 2: delete 'miscellaneous' which only has 3 data in dataset---
    # select data without 'miscellaneous'
    df_filtered = climate_df[climate_df['logical_fallacies'] != 'miscellaneous'].copy() 

    # --- step 3: randomly sample 10 data for each category in logical_fallacies ---
    
    # define a function to sample each group
    def custom_sample(group):
        # if the number of records in the group is less than or equal to 10, sample all records
        # otherwise, sample 10 records
        n_to_sample = min(len(group), SAMPLE_SIZE)
        return group.sample(n=n_to_sample, random_state=RANDOM_SEED)

    # conduct grouped sampling
    df_sampled = (
        df_filtered.groupby('logical_fallacies', dropna=False) # dropna=False 以保留 NaN 組
                   .apply(custom_sample)
                   .reset_index(drop=True)
    )

    # --- step 4: output to CSV file ---
    output_filename = 'climate_sampled_data.csv'
    df_sampled.to_csv(output_filename, index=False)

    print("--- Result ---")
    print(f"Original count: {len(climate_df)}")
    print(f"After delete 'miscellaneous': {len(df_filtered)}")
    print(f"Final sampled count: {len(df_sampled)}")
    print("\neach category count:")
    print(df_sampled['logical_fallacies'].value_counts(dropna=False).sort_index())
    
    # confirmation message
    print(f"\nsuccessfully generated file: {output_filename}")
```
